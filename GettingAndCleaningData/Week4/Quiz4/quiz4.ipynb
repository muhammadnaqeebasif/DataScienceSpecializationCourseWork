```r
gdp_dt <- fread(input = "https://d396qusza40orc.cloudfront.net/getdata%2Fdata%2FGDP.csv",
               skip=4,select = c(1,2,4,5), col.names = c("countrycode","rank","countryname","economy"),
               nrows=190)
gdp_dt
```

```r
edu_dt <- fread(input = "https://d396qusza40orc.cloudfront.net/getdata%2Fdata%2FEDSTATS_Country.csv")
setnames(edu_dt,names(edu_dt),tolower(gsub(" ","",names(edu_dt))))
```
